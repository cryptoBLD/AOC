```
#AOC 1.1
read = open("AOC 1.txt", "r")
read1 = read.read()
lines = read1.splitlines()
numbers = []
for i in lines:
  numbers.append(i)
for i in numbers:
  for p in numbers:
    if int(i) + int(p) == 2020:
      print(int(i)*int(p)) 
```

```
#AOC 1.2
read = open("AOC 1.txt", "r")
read1 = read.read()
lines = read1.splitlines()
numbers = []
for i in lines:
  numbers.append(i)
for i in numbers:
  for p in numbers:
    for l in numbers:
      if int(i) + int(p) + int(l) == 2020:
        print(int(i)*int(p)*int(l)) 
```
